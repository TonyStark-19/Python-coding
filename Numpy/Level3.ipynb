###  Practice with real world data

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 120000, 140000, 160000, 190000],
    [3, 200000, 230000, 260000, 300000],
    [4, 180000, 210000, 240000, 270000],
    [5, 160000, 185000, 205000, 230000]
])
sales_data
```

```python
# check shape
sales_data.shape
```

```python
# get first 3 rows
sales_data[:3]
```

```python
# total sales per year
np.sum(sales_data[:, 1:], axis=0)
```

```python
# Minimum sales per restaurant
np.min(sales_data[:, 1:], axis=1)
```

```python
# Maximum sales per year
np.max(sales_data[:, 1:], axis=0)
```

```python
# Average sales per restaurant (we can also use the average() function)
np.mean(sales_data[:, 1:], axis=1)
```

```python
# cumulative sales (adds previous values)
cf = np.cumsum(sales_data[:, 1:], axis=1)
cf
```

```python
# visualise it
plt.figure(figsize=(10, 6))
plt.plot(np.mean(cf, axis=0))
plt.title("Average cumulative sales per year across restaurants")
plt.xlabel("Years")
plt.grid()
plt.ylabel("Cumulative sales")
plt.show()
```

```python
# vectorise operations
restaurant_types = np.array(['biryani', 'chinese', 'pizza', 'burgur', 'cafe'])
vectorised_upper = np.vectorize(str.upper)
vectorised_upper(restaurant_types)
```
