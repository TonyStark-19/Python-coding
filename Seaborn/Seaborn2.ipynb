```python
# import libraries
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# load the penguins datasets
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# bar plot
sns.barplot(data = penguins, x = "species", y = "body_mass_g", hue = "sex",  estimator = np.mean)
```

```python
# countplot
sns.set_style("whitegrid")
sns.countplot(data = penguins, x = "species", hue = "island", palette = "Set2")
```

```python
# box plot
sns.boxplot(data = penguins, x = "species", y = "body_mass_g", hue = "island")
```

```python
# violin plot
sns.violinplot(data = penguins, x = "species", y = "body_mass_g", hue = "sex", split = False)
```

```python
# violin and swarm plot together
sns.violinplot(data = penguins, x = "species", y = "body_mass_g", color = "lightgreen")
sns.swarmplot(data = penguins, x = "species", y = "body_mass_g", hue = "island", size = 3, palette = "dark")
```

```python
# KDE plot - smooth version of histogram
sns.kdeplot(data = penguins, x = "body_mass_g", hue = "species", fill = True)
```

```python
# columns with numerical values
columns = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
# corelation
penguins[columns].corr()
```

```python
# Heatmap - works for numerical only
sns.heatmap(data = penguins[columns].corr(), annot = True, cmap = "Blues", vmin = -0.2, linewidths = "2", linecolor = "orange")
```

```python
# Rug plot
sns.rugplot(data = penguins, x = "body_mass_g", hue = "species", palette = "Set2", height = 0.2)
```

```python
# Pair plot
# defalut it takes numerical only, that's why we can skip columns
sns.pairplot(data = penguins, hue = "sex", palette = "Set2")
```

```python
# Pair grid: subplot grid for plotting pairwise relationships in a dataset
graph = sns.pairplot(data = penguins, hue = "sex", palette = "Set2")
# upper trianlge
graph.map_upper(sns.scatterplot)
# lower triangle
graph.map_lower(sns.kdeplot)
# diagornal
graph.map_diag(sns.histplot)
graph.add_legend()
```
