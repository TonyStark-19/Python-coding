```python
# import libraries
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# seaborn datasets
sns.get_dataset_names()
```

```python
# load the penguins datasets
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# find the number of species
penguins["species"].value_counts()
```

```python
# find the number of islands
penguins["island"].value_counts()
```

```python
# plot a scatter chart
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "island")
```

```python
# plot a scatter chart with set style
# white, dark, whitegrid, darkgrid, ticks
sns.set_style("white")
# set context (notebook for jn, paper is smaller, while talk and poster are for bigger representation)
sns.set_context("notebook")
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "sex", palette = "Dark2")
# despine to remove upper and right lines of graph
# pass a specific value like left = True for more customazation 
sns.despine()
```

### 

```python
# plot a scatter chart with style
# alpha for opacity
sns.scatterplot(data = penguins, x = "species", y = "body_mass_g", hue = "island", style = "sex", alpha = 0.5)
```

```python
# plot a striplot
# dodge for separating points on the bases of hue
sns.stripplot(data = penguins, x = "species", y = "body_mass_g", hue = "island", dodge = True, jitter = True)
```

```python
# swarm plot
sns.set_context("talk")
sns.swarmplot(data = penguins, x = "species", y = "body_mass_g", hue = "island")
sns.despine()
```

```python
# plot histogram
sns.set_context("notebook")
sns.histplot(data = penguins, x = "body_mass_g", hue = "sex", multiple = "stack")
```

```python
# regression plot
sns.regplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", scatter = True)
```

```python
# line plot
sns.set_style("whitegrid")
sns.lineplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", hue = "island", style = "sex")
```

```python
# joint plot
sns.jointplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm")
```

```python
# joint plot
sns.jointplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", hue = "island", kind = "scatter")
```
