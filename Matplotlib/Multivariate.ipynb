## Multivariate Analysis

```python
# import libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Salary" : [25000, 30000, 37000, 28000, 39000, 48000, 55000, 52000, 35000, 26000, 27000, 31000, 29000, 54000, 26000, 16000, 43000, 18000, 53000, 22000],
    "dept" : ["HR", "IT", "Finance", "HR", "Finance", "IT", "HR", "IT", "Finance", "HR"] * 2,
    "Age" : [22, 25, 29, 24, 30, 35, 40, 36, 28, 22, 23, 25, 26, 24, 31, 29, 27, 33, 35, 39],
    "experience" : [1, 2, 4, 1.5, 5, 8, 12, 10, 4, 1, 1.2, 2, 3, 1.8, 5.5, 4.5, 3.2, 7, 8, 10]
}
# convert data into a data frame
df = pd.DataFrame(data)
df.head()
```

### Multivariate analysis - 3 numerical columns

```python
# bubble plot
plt.scatter(df["Age"], df["Salary"], s = df["experience"]*20, color = "skyblue", edgecolors = "black")
plt.tight_layout()
plt.show()
```

### Multivariate analysis - 2 numerical and 1 catagorical column

```python
# scatter plot
color = {"HR": "yellow", "IT": "blue", "Finance": "green"}

for dept, color in color.items():
    df_dept = df[df["dept"] == dept]
    plt.scatter(df_dept["Age"], df_dept["Salary"], c = color, label = dept)

plt.legend()
plt.tight_layout()
plt.show()
```
