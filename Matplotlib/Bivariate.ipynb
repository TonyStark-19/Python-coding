## Bivariate Analysis

```python
# import libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Salary" : [25000, 30000, 37000, 28000, 39000, 48000, 55000, 52000, 35000, 26000, 27000, 31000, 29000, 54000, 26000, 16000, 43000, 18000, 53000, 22000],
    "dept" : ["HR", "IT", "Finance", "HR", "Finance", "IT", "HR", "IT", "Finance", "HR"] * 2,
    "Age" : [22, 25, 29, 24, 30, 35, 40, 36, 28, 22, 23, 25, 26, 24, 31, 29, 27, 33, 35, 39]
}
# convert data into a data frame
df = pd.DataFrame(data)
df.head()
```

### Bivariate analysis - numerical and numerical

```python
# scatter plot
plt.scatter(df["Age"], df["Salary"], color = "orange")
plt.tight_layout()
plt.show()
```

```python
# sort age before line chart plotting
sort_age = df.sort_values("Age")
```

```python
# Line chart
plt.plot(sort_age["Age"], df["Salary"], color="red", marker="o")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Bar chart
plt.bar(df["Age"], df["Salary"], color = "green")
plt.tight_layout()
plt.show()
```

### Bivariate analysis : numerical and catagorical

```python
# get HR salary
hr_sal = df[df["dept"] == "HR"]["Salary"]
hr_sal
```

```python
# get IT salary
it_sal = df[df["dept"] == "IT"]["Salary"]
it_sal
```

```python
# get Finance salary
fin_sal = df[df["dept"] == "Finance"]["Salary"]
fin_sal
```

```python
# box plot
plt.boxplot([hr_sal, it_sal, fin_sal], tick_labels = ["HR", "IT", "Finance"])
plt.tight_layout()
plt.show()
```

```python
# group salary count by department
salary_by_dept = df.groupby("dept")["Salary"].sum()
salary_by_dept
```

```python
# pie chart
plt.pie(salary_by_dept, labels = salary_by_dept.index, autopct = "%1.1f")
plt.tight_layout()
plt.show()
```

```python
# calculate the mean or average of salaries by department
hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
fin_mean = sum(fin_sal)/len(fin_sal)
```

```python
# bar plot
plt.bar(["HR", "IT", "Finance"], [hr_mean, it_mean, fin_mean], color = ["green", "red", "orange"])
plt.tight_layout()
plt.grid()
plt.show()
```
