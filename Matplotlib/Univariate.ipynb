## Univariate Analysis

```python
# import libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# basic data plotting
x = [1, 2, 3]
y = [4, 5, 6]
```

```python
# plot a line graph
plt.plot(x, y)
plt.grid()
plt.tight_layout()
plt.show()
```

### Univariate analysis - numerical

```python
data = {
    "Salary" : [25000, 30000, 37000, 28000, 39000, 48000, 55000, 52000, 35000, 26000, 27000, 31000, 29000, 54000, 26000, 16000, 43000, 18000, 53000, 22000]
}
# convert data into a data frame
df = pd.DataFrame(data)
df.head()
```

```python
# total records
df.shape
```

```python
# Line chart
plt.plot(df["Salary"], color="red", marker="o", linestyle=":", linewidth="2")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Histogram
plt.hist(df["Salary"], bins=5)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# box plot
plt.boxplot(df["Salary"])
plt.tight_layout()
plt.show()
```

### Univariate analysis - categorical

```python
# create new column: department
df["dept"] = ["HR", "IT", "Finance", "HR", "Finance", "IT", "HR", "IT", "Finance", "HR"] * 2 
```

```python
df.head()
```

```python
# count by values
count = df["dept"].value_counts()
count
```

```python
# Pie chart
plt.pie(count, labels = count.index, autopct = "%1.1f", explode = [0, 0.1, 0])
plt.tight_layout()
plt.axis("equal")
plt.show()
```

```python
# Countplot
plt.bar(count.index, count, color = ["green", "red", "orange"])
plt.tight_layout()
plt.show()
```
